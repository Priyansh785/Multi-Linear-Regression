```python
# Collinearity Problem Check
# Calculate VIF = 1/(1-Rsquare) for all independent variables

rsq_r=smf.ols("RD~ADM+MTS",data=df).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADM~RD+MTS",data=df).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MTS~RD+ADM",data=df).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RD','ADM','MTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```
